```python
import os
import shutil
from pathlib import Path

source_directory = Path(r"CONFIGURE SEU DIRETÓRIO AQUI")

pastas = {
    "Imagens": [".jpg", ".jpeg", ".png", ".gif", ".bmp", ".webp", ".svg", ".tiff"],
    "Videos": [".mp4", ".avi", ".mov", ".mkv", ".wmv", ".flv", ".webm", ".m4v", ".MOV"],
    "Documentos": [".pdf", ".doc", ".docx", ".xls", ".xlsx", ".ppt", ".pptx", ".txt", ".rtf", ".odt", ".csv", ".xlsm", ".xlsb"],
    "Compactados": [".zip", ".rar", ".7z", ".tar", ".gz", ".bz2"],
    "Executaveis": [".exe", ".msi", ".dmg", ".pkg", ".deb", ".rpm"],
    "Audio": [".mp3", ".wav", ".aac", ".flac", ".ogg", ".m4a", ".wma"],
    "Codigo": [".py", ".js", ".html", ".css", ".java", ".c", ".cpp", ".h", ".json", ".xml"]
}

for pasta, extensoes in pastas.items():
    pasta_path = source_directory / pasta
    pasta_path.mkdir(exist_ok=True)

    for arquivo in source_directory.iterdir():
        if arquivo.is_file() and arquivo.suffix in extensoes:
            arquivo.rename(pasta_path / arquivo.name)

```
